```python
def convert_columns_to_float(data):
    for column in data.columns:
        if data[column].dtype == object:
            if data[column].apply(lambda x: isinstance(x, (int, float))).all():
                data[column] = data[column].astype(float)
    return data
```

```python
def get_categorical_numerical_columns(data):
    categorical_columns = []
    numerical_columns = []

    for column in data.columns:
        if data[column].dtype == np.object:  # Check if the column type is object (categorical)
            categorical_columns.append(column)
        else:
            numerical_columns.append(column)

    return categorical_columns, numerical_columns
```

```python
categorical_cols_train, numerical_cols_train = get_categorical_numerical_columns(df_train)

print("Categorical columns:", categorical_cols_train)
print("Numerical columns:", numerical_cols_train)
```

## Encoding Categorical Data

```python
from sklearn.preprocessing import MultiLabelBinarizer

mlb = MultiLabelBinarizer()
encoded_data = mlb.fit_transform(df_train['packaging_materials'])
df_encoded_train = pd.DataFrame(encoded_data, columns=mlb.classes_)

substring = 'en:'
new_columns = [col.replace(substring, '') for col in df_encoded_train.columns]
df_encoded_train.columns = new_columns
new_columns.sort()
keep_columns = new_columns
df_encoded_train = df_encoded_train[new_columns]
numeric_data_train = numeric_data_train.reset_index(drop=True)

df_train_encoded = pd.concat([numeric_data_train, df_encoded_train], axis=1)
df_train_encoded
```

```python
print(keep_columns)
```

```python
mlb = MultiLabelBinarizer()
encoded_data = mlb.fit_transform(df_test['packaging_materials'])
df_encoded_test = pd.DataFrame(encoded_data, columns=mlb.classes_)

substring = 'en:'
new_columns = [col.replace(substring, '') for col in df_encoded_test.columns]
df_encoded_test.columns = new_columns
new_columns.sort()
df_encoded_test = df_encoded_test[new_columns]

print(new_columns)

for column in keep_columns:
    if column not in df_encoded_test.columns:
        df_encoded_test[column] = 0

df_encoded_test = df_encoded_test[keep_columns]

numeric_data_test = numeric_data_test.reset_index(drop=True)

df_test_encoded = pd.concat([numeric_data_test, df_encoded_test], axis=1)
df_test_encoded
```

## Separar las varaibles independientes y dependientes

```python
# Set independent and dependent variables

# Select Features
feature = df_train_encoded.drop('ecoscore_grade', axis=1)

# Select Target
target = df_train_encoded['ecoscore_grade']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(feature, target,
                                                    shuffle = True,
                                                    test_size=0.2,
                                                    random_state=1)

print('Shape of training feature:', X_train.shape)
print('Shape of testing feature:', X_test.shape)
print('Shape of training label:', y_train.shape)
print('Shape of training label:', y_test.shape)
```

```python
y_train
```

## Haciendo el modelo

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression(solver='saga', max_iter=150) # lbfgs / liblinear / sag / saga
model.fit(X_train, y_train)

accuracy = model.score(X_test, y_test)
print("Accuracy:", accuracy*100)
```
